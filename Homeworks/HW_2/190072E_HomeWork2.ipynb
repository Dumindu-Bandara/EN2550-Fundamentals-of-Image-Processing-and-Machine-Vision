```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Question 01

f = cv.imread("images/spider.png")
assert f is not None

gamma_list = [0.2, 0.8, 1.2, 2]

for gamma in gamma_list:
    t = np.array([(p/255)**gamma*255 for p in range(0,256)]).astype(np.uint8)
    g = cv.LUT(f,t)


    fig, ax = plt.subplots(1, 1)
    ax.plot(t)
    ax.set_title(f"Gamma = {str(gamma)}")

    fig, ax = plt.subplots(1, 1)
    plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
```

```python
t_1 = np.linspace(0, 100, 50, endpoint= True)
t_2 = np.linspace(100, 255, 150, endpoint = True)
t_3 = np.array([255 for p in range(200, 256)])

t = np.concatenate((t_1, t_2, t_3), axis = 0).astype(np.uint8)
fig, ax = plt.subplots()
ax.plot(t)
ax.set_aspect('equal')
assert len(t) == 256
g = cv.LUT(f, t)


fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))

fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
```

```python
f = cv.imread("images/shells.tif", cv.IMREAD_GRAYSCALE)

print(f.shape)

hist_f = cv.calcHist([f], [0], None, [256], [0,256])
g = cv.equalizeHist(f)
hist_g = cv.calcHist([g], [0], None, [256], [0,256])

fig, ax = plt.subplots(2, 1, figsize=(8,8))

ax[0].plot(hist_f)
ax[1].plot(hist_g)

fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))

fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))

# cv.namedWindow('Image', cv.WINDOW_AUTOSIZE)
# cv.imshow('Image', f)
# cv.waitKey(0)
# cv.imshow('Image', g)
# cv.waitKey(0)
# cv.destroyAllWindows()
```

```python
#Changing Saturation
f = cv.imread("images/zion_pass.jpg")
assert f is not None

hsv = cv.cvtColor(f, cv.COLOR_BGR2HSV)

Mask = cv.inRange(hsv, (0, 0, 0), (97, 100, 255))

hsv[:, :, 1] = Mask


fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
ax.set_title("Original")

fig, ax = plt.subplots(1, 1)
plt.imshow(cv.cvtColor(hsv, cv.COLOR_HSV2RGB))
ax.set_title("Processed Image")

```
